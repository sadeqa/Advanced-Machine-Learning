## GMM

A Gaussian mixture model is a probabilistic model that assumes all the data points are generated from a mixture of a finite number of Gaussian distributions with unknown parameters. After estimation of those parameters we get an estimation of the distribution of our data. For the clustering task, one can think of mixture models as generalizing k-means clustering to incorporate information about the covariance structure of the data as well as the centers of the latent Gaussians.

```python
class my_GMM():
    
    def __init__(self, k, initialization="kmeans"):
        '''
        Attributes:
        
        k_: integer
            number of components
        initialization_: {"kmeans", "random"}
            type of initialization
        mu_: np.array
            array containing means
        Sigma_: np.array
            array cointaining covariance matrix
        cond_prob_: (n, K) np.array
            conditional probabilities for all data points 
        labels_: (n, ) np.array
            labels for data points
        '''
        self.k_ = k
        self.initialization_ = initialization
        self.mu_ = None
        self.Sigma_ = None
        self.cond_prob_ = None
        self.proportions_k_ = None
        self.labels_ = None
        self.iterations_ = 0
        
    
    def compute_condition_prob_matrix(self, X, mu, Sigma, proportions_k):
            '''Compute the conditional probability matrix 
            shape: (n, K)
            '''
            n, _ = X.shape
            
            prob_z_x = np.zeros((n ,self.k_))
            
            for i in range(self.k_) : prob_z_x[:,i] = multivariate_normal.pdf(X, mu[i], Sigma[:,:,i])
            
            prob_z_x *= proportions_k 
            
            prob_z_x /= np.sum(prob_z_x, axis= 1, keepdims=True)
            
            return prob_z_x
            
            
    def compute_expectation(self, X, cond_prob):
        '''Compute the expectation to check increment'''
        
        n, p = X.shape

        prob_k_t = np.mean(cond_prob, axis=0)

        mu_t = np.dot(cond_prob.T, X)/(prob_k_t[:, None]*n)

        sigma_t = np.zeros((p,p,self.k_))
        
        for i in range(self.k_) : sigma_t[:,:,i] = np.dot((X - mu_t[i]).T, (cond_prob[:,[i]]*(X - mu_t[i])))/prob_k_t[i]

        sigma_t /= n

        return prob_k_t, mu_t, sigma_t

        
    def fit(self, X):
        """ Find the parameters mu_ and Sigma_
        that better fit the data
        
        Parameters:
        -----------
        X: (n, p) np.array
            Data matrix
        
        Returns:
        -----
        self
        """
        
        n,p = X.shape
        self.iterations_ = 0
                                         
                                        
        
        # TODO:
        # initialize the parameters
        # if initialize_ is kmeans, apply sklearn kmeans 
        # else randomly initialize them
        
        if self.initialization_ == "kmeans":
            
            model = KMeans(self.k_)
            model.fit(X)
            
            
            self.mu_ = model.cluster_centers_
            self.proportions_k_ = np.unique(model.labels_, return_counts=True)[1]/n

            self.Sigma_ = np.zeros((p,p,self.k_))
            for gaussian in range(self.k_):
                X_centered = X[model.labels_ == gaussian] - self.mu_[gaussian]
                self.Sigma_[:,:,gaussian] = np.dot(X_centered.T, X_centered)/X_centered.shape[0]

        else : 
            self.mu_ = np.random.rand(self.k_, p)
            self.proportions_k_ = np.ones(self.k_)/self.k_
            self.Sigma_ = np.repeat(np.eye(p)[:,:,None],self.k_, axis=2) 
        
            

        # While not(convergence)
        #     Compute conditional probability matrix
        #     Update parameters
                                                   
        converge = False
                                                   
        while not(converge):
                                                   
            old_params = {'mu' : self.mu_, 'sigma' : self.Sigma_, 'pi_k' : self.proportions_k_}
                     
            self.cond_prob_ = self.compute_condition_prob_matrix(X, self.mu_, self.Sigma_, self.proportions_k_)
                                                   
            self.proportions_k_ , self.mu_, self.Sigma_ = self.compute_expectation(X, self.cond_prob_)
            
            new_params = {'mu' : self.mu_, 'sigma' : self.Sigma_, 'pi_k' : self.proportions_k_}
            
            converge_mu = np.isclose(new_params['mu'], old_params['mu'], rtol= 1e-4, atol=0).all()
            
            converge_sigma =  np.isclose(new_params['sigma'], old_params['sigma'], rtol= 1e-4, atol=0).all()
            
            converge_prob_k = np.isclose(new_params['pi_k'], old_params['pi_k'], rtol= 1e-4, atol=0).all()
            
            converge = converge_mu*converge_sigma*converge_prob_k
            
            self.iterations_ += 1
            

        # Update labels_
        
        self.labels_ = np.argmax(self.cond_prob_, axis=1)

        # Return self
        return self
        
    def predict(self, X):
        """ Predict labels for X
        
        Parameters:
        -----------
        X: (n, p) np.array
            New data matrix
        
        Returns:
        -----
        label assigment        
        """
        # TODO
        return np.argmax(self.compute_condition_prob_matrix(X, self.mu_, self.Sigma_, self.proportions_k_), axis=1)
        
    def predict_proba(self, X):
        """ Predict probability vector for X
        
        Parameters:
        -----------
        X: (n, p) np.array
            New data matrix
        
        Returns:
        -----
        proba: (n, k) np.array        
        """
        # TODO
        return self.compute_condition_prob_matrix(X, self.mu_, self.Sigma_, self.proportions_k_)
```
